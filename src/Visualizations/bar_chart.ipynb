```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Metrics
metrics = ['Precision (Bullish)', 'Precision (Neutral)', 'Precision (Bearish)', 'Accuracy', 'F1-Score']

# Models and colors
models_order = ['CryptoBERT', 'Stocktwits-crypto (fine-tuned)', 'Stocktwits-crypto (baseline)', 'crypto_BiLSTM']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Scores dictionary
scores = {
    'CryptoBERT': [81.79, 65.55, 62.19, 75.01, 69.79],
    'Stocktwits-crypto (fine-tuned)': [80.01, 63.33, 64.42, 74.10, 68.39],
    'Stocktwits-crypto (baseline)': [70.30, 17.00, 25.34, 38.99, 33.11],
    'crypto_BiLSTM': [76.56, 58.48, 60.07, 70.95, 70.02]
}

# X positions
x = np.arange(len(metrics))
```

```python
# Adjust Y-axis limit back to 100

fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.2
opacity = 0.9

bars = []
for i, model in enumerate(models_order):
    bar = ax.bar(x + (i - 1.5) * bar_width, scores[model], bar_width, alpha=opacity, label=model, color=colors[i])
    bars.append(bar)

# Set axis labels and formatting
ax.set_xlabel('Metrics')
ax.set_ylabel('Score (%)')
ax.set_title('Performance Comparison of Cryptocurrency Sentiment Models')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.set_ylim(0, 100)
ax.legend()
ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.6)


# Annotate bars
for bar_group in bars:
    for bar in bar_group:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
